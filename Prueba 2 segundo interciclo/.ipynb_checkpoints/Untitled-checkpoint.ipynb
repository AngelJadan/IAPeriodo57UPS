```python
import csv
from neo4j import GraphDatabase
```

```python
text = open ('C:/Users/EstAngelMesiasJadanC/Desktop/claves.txt','r')
datos = (text.read()).split(sep=",")
user = datos[0]
pasw = datos[1]
print(user)
print("_____")
print(pasw)

uri = "neo4j://localhost:7687"
driver = GraphDatabase.driver(uri, auth=("neo4j", pasw))

def create_friend_of(tx, name, friend):
    tx.run("CREATE (a:Person)-[:KNOWS]->(f:Person {name: $friend}) "
           "WHERE a.name = $name "
           "RETURN f.name AS friend", name=name, friend=friend)

with driver.session() as session:
    session.write_transaction(create_friend_of, "Alice", "Bob")

with driver.session() as session:
    session.write_transaction(create_friend_of, "Alice", "Carl")

driver.close()
```
