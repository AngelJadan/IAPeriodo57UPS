```python
import requests
from bs4 import BeautifulSoup
import re
import csv
```

```python
cont=1
for i in range(3):
    print('Page')
    print(cont)
    url = 'https://www.ask.com/web?o=0&l=dir&qo=pagination&q=andres%20arauz%20noticias&qsrc=998&page='+str(cont)
    respuesta = requests.get(
        url,
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Safari/537.36'
        }
    )
    contenido_web = BeautifulSoup(respuesta.text, 'lxml')
    noticias = contenido_web.find('div', attrs={'PartialSearchResults-body'})
    seccion = noticias.findChildren('div', attrs={'class': 'PartialSearchResults-item'})
    
    for item in seccion:
        print("Item")
        
        tit = str(item.find('div').a)
        titulo = BeautifulSoup(tit, "lxml").text
        print("Pagina: "+titulo)
        
        link = str(item.find('div').a.get("href"))
        print("Link: "+link)
        
        sms = str(item.find('p', attrs = {'class':'PartialSearchResults-item-abstract'}))
        sms1 = BeautifulSoup(sms, "lxml").text
        print("Mensaje: "+sms1)
        
    cont=cont+1
```
