```python
import requests
from bs4 import BeautifulSoup
import re
import csv
```

```python
cont=0
for i in range(5):
    print('Item')
    print(cont)
    url = 'https://www.ecosia.org/news?q=Andres%20Arauz&p='+str(cont)
    respuesta = requests.get(
        url,
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Safari/537.36'
        }
    )
    myData = []
    contenido_web = BeautifulSoup(respuesta.text, 'lxml')
    noticias = contenido_web.find('div', attrs={'class':'layout'})
    sections = noticias.findChildren('div', attrs={'class':'result__body'})
    c=0
    for section in sections:
        fila = []
        #Codigo para grafo
        codigo = "unesecosia"+str(c)
        #print('Link:')
        link = section.a.get('href')
        clink = BeautifulSoup(str(link), "lxml").text
        
        #Link limpio en variable
        clink = str(clink)
               
        hora = section.findChildren('time', attrs={'class':'news-result__date'})
        chora = BeautifulSoup(str(hora), "lxml").text
        #Hora limpia en variable
        schora = (chora.strip("[").strip("]")).rstrip().lstrip()
    
        sms = section.findChildren('div', attrs={'class':'result__content-inner'})
        csms = BeautifulSoup(str(sms), "lxml").text
        
        #Mensaje limpio en variable
        scsms =(csms.strip("[").strip("]").strip('"')).rstrip().lstrip()
        #print(scsms)
        
        #Inserccion en arreglo de filas
        fila.append(codigo)
        fila.append(str(clink))
        fila.append(str(schora))
        fila.append(str(scsms))
        #Inserccion en arreglo para csv
        myData.append(fila)
        
        #print("Link")
        #print(str(clink))
        #print("Hora:")
        #print(schora.rstrip().lstrip())
        #print("Mensaje:")
        #print(scsms)
        #print('******')
        c=c+1
    myFile = open('candPresEcosia'+str(cont)+str(c)+'.csv', 'w')
    with myFile:
        writer = csv.writer(myFile)
        writer.writerows(myData)
    print("Archivo guardado"+str(cont))
    cont=cont+1
```
