```python
from neomodel import (config, StructuredNode, StringProperty, IntegerProperty,
    UniqueIdProperty, RelationshipTo, RelationshipFrom)

config.DATABASE_URL = 'bolt://neo4j:Angel2019@localhost:7687'

class Asambleista(StructuredNode):
    nameCan = StringProperty(unique_index=True)
    partido = StringProperty(unique_index=True)
class Presidente(StructuredNode)  :
    namePresi = StringProperty(unique_index = True)
    partPresi = StringProperty(unique_index = True)
class Noticia(StructuredNode):
    idNoti = UniqueIdProperty()
    date = StringProperty(unique_index=False)
    messaje = StringProperty(unique_index=False)
    link = StringProperty(unique_index=False)
    asambleista = RelationshipTo('Asambleista' , 'NOTICIA_DE')
    presidente = RelationshipTo
```
